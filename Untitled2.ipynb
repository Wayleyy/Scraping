```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la page de Tesla pour les revenus
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# Ajouter des en-tÃªtes pour simuler un navigateur
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
}

# Envoyer la requÃªte HTTP avec les en-tÃªtes
response = requests.get(url, headers=headers)

# VÃ©rifier si la requÃªte a rÃ©ussi
if response.status_code == 200:
    # Parser la page HTML avec BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Chercher le tableau avec les donnÃ©es financiÃ¨res
    table = soup.find('table', {'class': 'historical_data_table'})

    if table:
        # Extraire les lignes du tableau
        rows = table.find_all('tr')

        # CrÃ©er une liste pour stocker les donnÃ©es
        data = []

        # Extraire les donnÃ©es ligne par ligne
        for row in rows[1:]:  # Ignorer la premiÃ¨re ligne (entÃªte)
            cols = row.find_all('td')
            if len(cols) > 0:
                year = cols[0].text.strip()
                revenue = cols[1].text.strip()
                data.append([year, revenue])

        # CrÃ©er un DataFrame Ã  partir des donnÃ©es extraites
        tesla_revenue = pd.DataFrame(data, columns=['Year', 'Revenue'])

        # Afficher les cinq derniÃ¨res lignes du DataFrame
        print(tesla_revenue.tail())
    else:
        print("Table not found on the page.")
else:
    print(f"Failed to retrieve the page, status code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la page web contenant les informations financiÃ¨res de GameStop
url = 'https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue'

# Ajouter un User-Agent pour simuler une requÃªte depuis un navigateur
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
}

# Envoyer une requÃªte GET avec les headers
response = requests.get(url, headers=headers)

# VÃ©rifier que la requÃªte a rÃ©ussi (code 200)
if response.status_code == 200:
    print("Page rÃ©cupÃ©rÃ©e avec succÃ¨s!")
else:
    print(f"Ãchec de la rÃ©cupÃ©ration de la page. Status code : {response.status_code}")

# Utiliser BeautifulSoup pour parser le contenu HTML
soup = BeautifulSoup(response.content, 'html.parser')

# Affichage de la structure HTML pour le dÃ©bogage
print(soup.prettify()[:1000])  # Afficher les 1000 premiers caractÃ¨res pour inspecter

# Trouver la table contenant les revenus (modifiez le sÃ©lecteur si nÃ©cessaire)
table = soup.find('table', {'class': 'historical_data_table'})

# VÃ©rifier si la table est trouvÃ©e
if table:
    print("Table trouvÃ©e!")
else:
    print("Table non trouvÃ©e. VÃ©rifiez la structure HTML de la page.")

# Si la table est trouvÃ©e, extraire les donnÃ©es
if table:
    data = []
    for row in table.find_all('tr')[1:]:  # Ignorer la premiÃ¨re ligne (en-tÃªtes)
        cols = row.find_all('td')
        if len(cols) > 1:
            year = cols[0].text.strip()
            revenue = cols[1].text.strip().replace('$', '').replace(',', '')  # Nettoyer les valeurs
            data.append([year, revenue])

    # CrÃ©er un DataFrame Ã  partir des donnÃ©es extraites
    gme_revenue = pd.DataFrame(data, columns=['Year', 'Revenue'])

    # Afficher les cinq derniÃ¨res lignes du DataFrame
    print(gme_revenue.tail())

    # Sauvegarder le DataFrame dans un fichier CSV
    gme_revenue.to_csv('gme_revenue.csv', index=False)
else:
    print("Impossible d'extraire les donnÃ©es. VÃ©rifiez la table HTML.")
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# TÃ©lÃ©charger les donnÃ©es boursiÃ¨res de Tesla avec yfinance
tesla_data = yf.download('TSLA', period='max')

# Fonction pour tracer un graphique de l'action
def make_graph(data):
    plt.figure(figsize=(10,6))
    plt.plot(data['Close'], label='Tesla Closing Price', color='blue')  # Tracer les prix de clÃ´ture
    plt.title('Tesla Stock Price (Closing)', fontsize=16)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Price (USD)', fontsize=12)
    plt.legend()
    plt.grid(True)
    plt.show()

# Appeler la fonction pour tracer le graphique
make_graph(tesla_data)
```
